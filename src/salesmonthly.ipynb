```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Read and prepare data
df = pd.read_csv('salesmonthly.csv', parse_dates=['datum'])
medicines = df.columns[1:]
colors = cm.tab10.colors  # Get 10 distinct colors

# 1. Individual Subplots (Original)
fig1, axes1 = plt.subplots(4, 2, figsize=(20, 20), num='Individual Trends')
axes1 = axes1.flatten()

for idx, med in enumerate(medicines):
    ax = axes1[idx]
    df.plot(x='datum', y=med, ax=ax, color=colors[idx], linewidth=1.5)
    ax.set_title(f'{med} Sales', fontsize=12)
    ax.tick_params(axis='x', rotation=45, labelsize=8)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend().remove()

plt.tight_layout(pad=3.0)

# 2. Combined Trend Plot (New)
plt.figure(figsize=(16, 8), num='Combined Trend Comparison')

for idx, med in enumerate(medicines):
    plt.plot(df['datum'], df[med],
             color=colors[idx],
             linewidth=1.5,
             alpha=0.8,
             label=med)

plt.title('All Medicine Sales Trends Comparison (2014-2019)', fontsize=14, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales Quantity', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

# Create legend outside plot with 2 columns
plt.legend(bbox_to_anchor=(1.02, 1),
           loc='upper left',
           borderaxespad=0,
           ncol=2,
           fontsize=10)

plt.tight_layout()
plt.show()
```
