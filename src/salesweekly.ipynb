```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Read and prepare weekly data
df = pd.read_csv('salesweekly.csv', parse_dates=['datum'], dayfirst=False)
medicines = df.columns[1:]
colors = cm.tab10.colors  # Get distinct colors

# 1. Individual Subplots (4x2 grid)
fig1, axes1 = plt.subplots(4, 2, figsize=(22, 20), num='Weekly Individual Trends')
axes1 = axes1.flatten()

for idx, med in enumerate(medicines):
    ax = axes1[idx]
    df.plot(x='datum', y=med, ax=ax, color=colors[idx], linewidth=1.2)
    ax.set_title(f'{med} Weekly Sales', fontsize=12, pad=12)
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=45, labelsize=8)
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend().remove()

plt.tight_layout(pad=3.0)

# 2. Combined Trend Plot
plt.figure(figsize=(18, 9), num='Weekly Combined Trends Comparison')

for idx, med in enumerate(medicines):
    plt.plot(df['datum'], df[med],
             color=colors[idx],
             linewidth=1.0,
             alpha=0.7,
             label=med)

plt.title('Weekly Medicine Sales Trends Comparison (2014-2019)', fontsize=14, pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales Quantity', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.grid(True, linestyle=':', alpha=0.4)

# Create organized legend
plt.legend(bbox_to_anchor=(1.02, 1),
           loc='upper left',
           borderaxespad=0.3,
           ncol=2,
           fontsize=10,
           title='Medicine Codes',
           title_fontsize=11)

plt.tight_layout()
plt.show()
```
